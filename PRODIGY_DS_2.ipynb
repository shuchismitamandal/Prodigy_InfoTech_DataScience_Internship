## **2. Data Cleaning**


```python
df = df.dropna(axis=1, how="all")
df.info()
```

```python
df = df.drop(columns=["Country Code", "Indicator Code"])
```

```python
df_long = df.melt(
    id_vars=["Country Name", "Indicator Name"],
    var_name="Year",
    value_name="Value"
)
df_long
```

```python
df_long.info()
```

```python
df_long["Year"] = pd.to_numeric(df_long["Year"], errors="coerce")
df_long.info()
```

## **3. Exploratory Data Analysis**


```python
# Count of unique indicators
print(df_long["Indicator Name"].nunique())
print("\n")

# See first 15 indicators 
print(df_long["Indicator Name"].unique()[:15])
```

```python
gdp = df_long[df_long["Indicator Name"] == "GDP (current US$)"]
population =df_long[df_long["Indicator Name"] == "Population, total"]
life_expectancy = df_long[df_long["Indicator Name"] == "Life expectancy at birth, total (years)"]
unemployment = df_long[df_long["Indicator Name"] == "Unemployment, total (% of total labor force) (modeled ILO estimate)"]
```

```python
import matplotlib.ticker as ticker

ax = gdp.plot(x="Year", y="Value", title= "What is the trend of India’s GDP growth over the decades?", legend=False)

# Axis labels
ax.set_xlabel("Year")
ax.set_ylabel("GDP (US$ Trillions)")

# Format y-axis to trillions
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f'{x/1e12:.1f} T')
)

plt.show()
```

```python
# Population trend
ax = population.plot(x="Year", y="Value", title= "How has India’s population grown since 1960?", legend=False)

# Axis labels
ax.set_xlabel("Year")
ax.set_ylabel("Population (Billions)")

# Format y-axis to trillions
ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f'{x/1e9:.1f} B')
)

plt.show()
```

```python
pop_growth = ((population["Value"].dropna().iloc[-1] - population["Value"].dropna().iloc[0])
              / population["Value"].dropna().iloc[0]) * 100

gdp_growth = ((gdp["Value"].dropna().iloc[-1] - gdp["Value"].dropna().iloc[0])
              / gdp["Value"].dropna().iloc[0]) * 100
print(f"Population Growth (1960-2020): {pop_growth:.2f}%")
print(f"GDP Growth (1960-2020): {gdp_growth:.2f}%")
```

```python
ax = life_expectancy.plot(x="Year", y="Value", title= "How has life expectancy in India changed over time?", legend=False)

# Axis labels
ax.set_xlabel("Year")
ax.set_ylabel("Age (Years)")

plt.show()
```

```python
ax = unemployment.plot(x="Year", y="Value", title= "What trends can we see in India’s unemployment rate since 1990?", legend=False)

# Axis labels
ax.set_xlabel("Year")
ax.set_ylabel("Unemployment Rate (%)")

plt.show()
```

```python
fig, ax1 = plt.subplots()

# Plot GDP
ax1.plot(gdp["Year"], gdp["Value"], color="blue", label="GDP (US$)")
ax1.set_xlabel("Year")
ax1.set_ylabel("GDP (US$) in Trillions")

# Create secondary y-axis for Population
ax2 = ax1.twinx()
ax2.plot(population["Year"], population["Value"], color="green", label="Population")
ax2.set_ylabel("Population in Billions")

# Title
plt.title("India's GDP vs Population (1960–2020)")

fig.legend()


plt.show()
```

```python
gdp1 = gdp[gdp["Year"] >= 1990]
unemployment = unemployment[unemployment["Year"] >= 1990]
```

```python
fig, ax1 = plt.subplots()

# GDP
ax1.plot(gdp1["Year"], gdp1["Value"]/1e12, color="teal", label="GDP")
ax1.set_xlabel("Year")
ax1.set_ylabel("GDP (Trillions USD)")

# Unemployment
ax2 = ax1.twinx()
ax2.plot(unemployment["Year"], unemployment["Value"], color="purple", label="Unemployment")
ax2.set_ylabel("Unemployment (%)")

plt.title("GDP vs Unemployment in India")
fig.legend()
plt.show()
```

```python
pop_growth = population["Value"].pct_change() * 100
life_growth = life_expectancy["Value"].pct_change() * 100

plt.plot(population["Year"], pop_growth, label="Population Growth (%)", color="blue")
plt.plot(life_expectancy["Year"], life_growth, label="Life Expectancy Growth (%)", color="green")
plt.xlabel("Year")
plt.title("Growth Rates: Population vs Life Expectancy")
plt.legend()
plt.show()
```
